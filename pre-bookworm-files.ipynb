```python
class Count_Array():
    def __init__(self, fn, grams = 1):
        self.path = Path(fn)
        self.fsize = p.stat().st_size
        self.nrow = int(self.fsize/(4 + 4 + grams * 4))
        self.mat = np.memmap(self.path, dtype='int32', mode='r+', shape=(self.nrow, 2 + grams))
    def sort(self):
        self.mat.view('i4, i4, i4').sort(order=['f0'], axis=0)
    def bisect_left(self, x, lo = 0, hi = None):
        if hi is None:
            hi = self.mat.shape[0]
        max = hi
        cap = hi
        while lo < hi:
            mid = (lo + hi) // 2
            if x < self.mat[mid][0]:
                hi = mid
            else:
                lo = mid + 1
        return lo
    def counts(self, x):
        left = self.bisect_left(x-1, 0, None)
        right = self.bisect_left(x, left, None)
#        print(left, right)
        return self.mat[left:right,1:]
    def countframe(self, ix):
        c = self.counts(ix)
        return pd.DataFrame(c[:,1], columns=["count"], index=c[:,0])
    def all_counts(self, ixes):
        c = self.countframe(ixes.pop())
        while len(ixes) > 0:\
            c += self.countframe(ixes.pop())
        return c
    
g = Count_array(p)
g.all_counts([2, 10]).sum()
```
